```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from pytorch_tabnet.tab_model import TabNetRegressor
import xgboost as xgb
import lightgbm as lgb
import torch
```

```python
# âœ… ë°ì´í„° ë¡œë”©
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')

# âœ… ê¸°ë³¸ ì „ì²˜ë¦¬
train.drop(columns=['ID'], inplace=True)
test.drop(columns=['ID'], inplace=True)

# ê¸°ì—…ê°€ì¹˜ ìˆ˜ì¹˜í™”

def parse_value_range(value):
    try:
        return float(value)
    except:
        if isinstance(value, str) and 'ì´ìƒ' in value:
            return float(value.replace('ì´ìƒ', '').strip())
        return np.nan

# ë³€ìˆ˜ ë¶„ë¥˜
category_features = ['ì„¤ë¦½ì—°ë„','êµ­ê°€','ë¶„ì•¼','íˆ¬ìë‹¨ê³„','ê¸°ì—…ê°€ì¹˜(ë°±ì–µì›)']
numeric_features = ['ì§ì› ìˆ˜','ê³ ê°ìˆ˜(ë°±ë§Œëª…)','ì´ íˆ¬ìê¸ˆ(ì–µì›)','ì—°ë§¤ì¶œ(ì–µì›)','SNS íŒ”ë¡œì›Œ ìˆ˜(ë°±ë§Œëª…)']
bool_features = ['ì¸ìˆ˜ì—¬ë¶€','ìƒì¥ì—¬ë¶€']

# ë²”ì£¼í˜• ì²˜ë¦¬
train['ì„¤ë¦½ì—°ë„'] = train['ì„¤ë¦½ì—°ë„'].astype('object')
test['ì„¤ë¦½ì—°ë„'] = test['ì„¤ë¦½ì—°ë„'].astype('object')

encoders = {}
for feature in category_features:
    encoders[feature] = LabelEncoder()
    train[feature] = train[feature].fillna('Missing').astype(str)
    test[feature] = test[feature].fillna('Missing').astype(str)
    train[feature] = encoders[feature].fit_transform(train[feature])
    test[feature] = encoders[feature].transform(test[feature])

# ë¶ˆë¦¬ì–¸ ì²˜ë¦¬
bool_map = {'Yes': 1, 'No': 0}
for feature in bool_features:
    train[feature] = train[feature].map(bool_map)
    test[feature] = test[feature].map(bool_map)

# ê¸°ì—…ê°€ì¹˜ ìˆ˜ì¹˜í™”
train['ê¸°ì—…ê°€ì¹˜(ì¤‘ê°„ê°’)'] = train['ê¸°ì—…ê°€ì¹˜(ë°±ì–µì›)'].apply(parse_value_range)
test['ê¸°ì—…ê°€ì¹˜(ì¤‘ê°„ê°’)'] = test['ê¸°ì—…ê°€ì¹˜(ë°±ì–µì›)'].apply(parse_value_range)

# ìˆ˜ì¹˜í˜• ê²°ì¸¡ì¹˜ ì²˜ë¦¬
all_numeric_features = numeric_features + ['ê¸°ì—…ê°€ì¹˜(ì¤‘ê°„ê°’)']
for feature in all_numeric_features:
    mean_value = train[feature].mean()
    train[feature] = train[feature].fillna(mean_value)
    test[feature] = test[feature].fillna(mean_value)

# íŒŒìƒ ë³€ìˆ˜ ìƒì„±
def create_derived_features(df):
    df = df.copy()
    df['ì§ì› 1ì¸ë‹¹ ë§¤ì¶œ'] = df['ì—°ë§¤ì¶œ(ì–µì›)'] / df['ì§ì› ìˆ˜']
    df['ê³ ê° 1ì¸ë‹¹ ë§¤ì¶œ'] = df['ì—°ë§¤ì¶œ(ì–µì›)'] / df['ê³ ê°ìˆ˜(ë°±ë§Œëª…)']
    df['ê³ ê° 1ì¸ë‹¹ íˆ¬ìê¸ˆ'] = df['ì´ íˆ¬ìê¸ˆ(ì–µì›)'] / df['ê³ ê°ìˆ˜(ë°±ë§Œëª…)']
    df['íˆ¬ì ëŒ€ë¹„ ê°€ì¹˜'] = df['ê¸°ì—…ê°€ì¹˜(ì¤‘ê°„ê°’)'] / df['ì´ íˆ¬ìê¸ˆ(ì–µì›)']
    df['SNSë¹„ìœ¨'] = df['SNS íŒ”ë¡œì›Œ ìˆ˜(ë°±ë§Œëª…)'] / df['ê³ ê°ìˆ˜(ë°±ë§Œëª…)']

    # ì¶”ê°€ ë³€ìˆ˜
    df['íˆ¬ì ëŒ€ë¹„ ì—°ë§¤ì¶œ'] = df['ì—°ë§¤ì¶œ(ì–µì›)'] / (df['ì´ íˆ¬ìê¸ˆ(ì–µì›)'] + 1e-3)
    df['ê°€ì¹˜ ëŒ€ë¹„ ì—°ë§¤ì¶œ'] = df['ì—°ë§¤ì¶œ(ì–µì›)'] / (df['ê¸°ì—…ê°€ì¹˜(ì¤‘ê°„ê°’)'] + 1e-3)
    df['ì—°ë§¤ì¶œ ì„±ì¥ ì ì¬ë ¥'] = df['SNS íŒ”ë¡œì›Œ ìˆ˜(ë°±ë§Œëª…)'] * df['ê³ ê°ìˆ˜(ë°±ë§Œëª…)']
    df.fillna(0, inplace=True)
    return df
```

```python
# íŒŒìƒ ë³€ìˆ˜ ìƒì„± (ì´ë¯¸ ì •ì˜ëœ í•¨ìˆ˜ ì‚¬ìš©)
train = create_derived_features(train)
test = create_derived_features(test)

# âœ… ê³µë°± ì œê±°
train.columns = train.columns.str.replace(' ', '_')
test.columns = test.columns.str.replace(' ', '_')
```

```python
# âœ… X, y ì¤€ë¹„
features = [col for col in train.columns if col != 'ì„±ê³µí™•ë¥ ']
X = train[features]
y = train['ì„±ê³µí™•ë¥ ']

# âœ… ì•™ìƒë¸” ì¤€ë¹„
N_FOLDS = 5
kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

# í…ŒìŠ¤íŠ¸ì…‹ ì˜ˆì¸¡ ì €ì¥
tabnet_test_preds = np.zeros((test.shape[0], N_FOLDS))
xgb_test_preds = np.zeros((test.shape[0], N_FOLDS))
lgb_test_preds = np.zeros((test.shape[0], N_FOLDS))

for fold, (train_idx, valid_idx) in enumerate(kf.split(X)):
    print(f"\nğŸ” Fold {fold+1}/{N_FOLDS}")
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    # TabNet
    tabnet = TabNetRegressor(
        n_d=32,
        n_a=32,
        n_steps=5,
        gamma=1.5,
        lambda_sparse=1e-4,
        optimizer_fn=torch.optim.AdamW,
        optimizer_params=dict(lr=2e-2),
        seed=42,
        verbose=0
    )

    tabnet.fit(X_train.values, y_train.values.reshape(-1, 1),
               eval_set=[(X_valid.values, y_valid.values.reshape(-1, 1))],
               patience=10, max_epochs=200)
    tabnet_test_preds[:, fold] = tabnet.predict(test.values).squeeze()

    # XGBoost
    xgb_model = xgb.XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )
    xgb_model.fit(X_train, y_train)
    xgb_test_preds[:, fold] = xgb_model.predict(test)

    # LightGBM
    lgb_model = lgb.LGBMRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )
    lgb_model.fit(X_train, y_train)
    lgb_test_preds[:, fold] = lgb_model.predict(test)

# âœ… ì„¸ ëª¨ë¸ ì˜ˆì¸¡ í‰ê·  (Soft Voting Ensemble)
final_preds = (
    tabnet_test_preds.mean(axis=1) +
    xgb_test_preds.mean(axis=1) +
    lgb_test_preds.mean(axis=1)
) / 3

# âœ… ì œì¶œ
sample_submission['ì„±ê³µí™•ë¥ '] = final_preds
sample_submission.to_csv('simple_ensemble_submission.csv', index=False, encoding='utf-8-sig')

print("\nâœ… ë‹¨ìˆœ ì•™ìƒë¸”(í‰ê· ) ì™„ë£Œ ë° ì œì¶œ íŒŒì¼ ì €ì¥ë¨!")
```
